# Wilson and Cowan Model (Dimensionless)

## No Drive

Generate the NeuroML/LEMS file with the Wilson and Cowan definition

```python
# need to move to NeuroML directory in order to do run the python script
% cd ../NeuroML2
! python GenerateWilsonCowanLEMS.py -ie0 0 -ii0 0 -isocline
```

Run pynml to simulate the defined LEMS

```python
import matplotlib.pyplot as plt
%matplotlib inline  
from pyneuroml import pynml

no_drive_lems_file='LEMS_WC_slowDL.xml'
no_drive_results = pynml.run_lems_with_jneuroml(no_drive_lems_file, nogui=True, load_saved_data=True, plot=False)
```

```python
no_drive_results.keys()
```

```python
plt.plot(no_drive_results['t'], no_drive_results['ExcPop/0/Exc/r'], label='E', color='r')
plt.plot(no_drive_results['t'], no_drive_results['InhPop/0/Inh/r'], label='I', color='b')
plt.xlabel('Time (ms)')
plt.ylabel('Proportion of firing cells')
plt.legend()
plt.grid(True)
plt.margins(0)
```
